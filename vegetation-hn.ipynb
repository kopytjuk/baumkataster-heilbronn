```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from matplotlib import cm
```

```python
file_path = "data/Allee.csv"
```

```python
df = pd.read_csv(file_path, encoding="cp1252", sep=";")
```

```python
df.head()
```

```python
df.columns
```

```python
len(df)
```

```python
tree_counts = df["Deutsche Bezeichnung"].value_counts()
```

```python
def format_ratio(r):
    return "{:d} ({:.1f}%)".format(int((r/100)*len(df)), r)
```

```python
fig, ax = plt.subplots(figsize=(20,20))
ax.set_title("Baumarten (deutsche Bez.) in der Allee, 74072 Heilbronn", fontsize=20)

wedges, texts, pct = ax.pie(tree_counts.values, labels=tree_counts.index, autopct=format_ratio, shadow=False, textprops={"fontsize": 17}, labeldistance=1.1, colors=cm.Paired.colors[:len(tree_counts)])
```
